# Using PangaeaDataset in PyeloTUPS

This tutorial demonstrates how to use `PangaeaDataset` to discover, explore,
and retrieve paleoceanographic datasets from PANGAEA.

Unlike `NOAADataset`, which interacts with the NOAA NCEI Paleo API,
`PangaeaDataset` communicates with the PANGAEA backend via `pangaeapy`,
while exposing a similar researcher-facing API.

### Requirements

Installation of ```pyleotups [version > {0.0.1}]``` automatically handles the installation of external services required for accessing the Pangaea Datasets. 


In this tutorial, we will cover:

0. **Credentials setup** [Optional] Setting up credentials to access private datasets
1.  **Searching** for studies using the Pangaea.
2.  **Browsing** search results (Publications, Geography, Funding).
3.  **Extracting** data into pandas DataFrames.

### 0. Credential Setup

Few datasets over Panagaea are protected and reuqires the user to have an ACCESS KEY. 

Follow the instructions to access and set the key through pyleotups utilities. You can also do the steps manually.

<details>
<summary><strong>1. Obtaining Your PANGAEA API Token</summary></strong>

1. Create an account or log in at:  
   https://www.pangaea.de/user/login.php  

2. After logging in, navigate to your PANGAEA profile.

3. Locate your **API login token key**.

4. Copy this token — it is unique to your account.

> Keep this key private. It grants access to datasets associated with your account and other personal information.


<details>
<summary><strong>2. Securely Storing Your Token</summary></strong>

To prevent accidental exposure of your API key in notebooks or version control, we recommend storing it in a local `.env` file.

PyeloTUPS provides a helper utility to securely manage this for you.



#### Step 1 — Save Your Credentials (One-Time Setup)

```python

save_pangaea_credentials("your_token_here")
```

#### Step 2 — Load Credentials in Your Notebook

```python

pan_api = load_pangaea_credentials()
```
The variable pan_api now securely contains your login token.

Installation of `pyleotups` (version > 0.0.1) automatically installs the external dependencies required to access PANGAEA datasets.

However, some datasets on PANGAEA are **protected** and require an **ACCESS KEY** (login token) for authentication.



#### Obtaining Your PANGAEA Access Key

1. Create an account or log in (for existing users) at:
   [https://www.pangaea.de/user/login.php](https://www.pangaea.de/user/login.php)

2. After logging in, navigate to your **PANGAEA profile**.

3. Locate your **login token key** in your profile settings.

> This token is unique to each user and should be kept private. 


### Securely Storing Your Token Using `dotenv`

To protect your personal credentials and prevent accidental exposure (e.g., when sharing notebooks), we strongly recommend using a `.env` file.

#### Step 1: Create a .env File

* In your project’s root directory, create a file named exactly:

```
.env
```

#### Step 2: Add Your API Token

Inside the .env file, add the following line:

```
PANGAEA_API="{your pangaea login token key}"
```

> If you use source control, we strongly recommend adding .env to your .gitignore files to exclude the .env file from being commited.
 
**Important:**

* The key must be enclosed in **double quotes**.
* Do **not** share this `.env` file when distributing notebooks or code.

---

### Accessing the Token in Your Notebook

In your Python notebook or script, load the token as follows:

```python
from dotenv import load_dotenv
import os
load_dotenv()
pan_api = os.getenv("PANGAEA_API")
```

The variable `pan_api` now securely contains your PANGAEA login token and can be used for accessing protected datasets.

### 1. Searching for Studies


````python
import pyleotups as pt

# Initialize a Pangaea Dataset manager
# This object will act as your session, storing search results and processing data.
dataset = pt.PangaeaDataset(auth_token=pan_api)
````


The `search_studies()` method queries the PANGAEA backend
and registers matching datasets internally.

Following parameters can be used to tune the search
- `q` → free-text query  
- `bbox` → geographic bounding box (minlon, minlat, maxlon, maxlat)  
- `keywords` → optional provider-specific terms 

````python
ds = dataset.search_studies(
    q="Khider, D.",
    # bbox=(-10, -10, 10, 10),  #(min_lon, min_lat, max_lon, max_lat) 
    # keywords=["Sr/Ca"],
    limit=10
)
````

### 2. EXPLORING RESULTS 

#### a. Getting a summary. 

You can access the search results in two ways:
- `dataset.get_summary()` → returns a DataFrame of search results (default behavior)
- ACCESS the DataFrame in return value. i.e. `ds`

All results are stored internally in `dataset.studies`

````python
dataset.get_summary()
````

#### b. Exploring Geographic Metadata

Returns site-level geographic information where available.

If no geospatial metadata exists, an empty DataFrame is returned.

````python
geo = dataset.get_geo()
display(geo)
````

#### c. Explore the listed Publication

````python
pubs = dataset.get_publications()
display(pubs)
````
